```python
## Grafico Ternário (Citoplasma)
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

 
#plt.rcdefaults()
fig, axs = plt.subplots(figsize=(1.2,3), sharey=True)
plt.style.use('ggplot')
#plt.style.use('seaborn')

fs = 11  # fontsize title
fs_ticks = 7  # fontsize  
fs_x = 9
fs_y = 10

plt.rcParams['font.family'] = 'Calibri'
#plt.rcParams['font.family'] = 'sans-serif'
#plt.rcParams['font.sans-serif'] = ['Tahome', 'DejaVu Sans',
#                               'Lucida Grande', 'Verdana']

#  data
features = ('circularity', 'area', 'fractal dim.', 'compacity', 'perimeter', 'area ratio',
            'equi. diameter', 'EFD 1', 'minor axis', 'solidity', 'EFD 28', 'aver. radial dist.', 'convex hull')
scores = [1.0955058e-01, 8.0719665e-02, 6.6231780e-02, 5.4855734e-02, 3.2158550e-02,
          1.5500017e-02, 1.3882077e-02, 6.6992515e-03, 5.4590139e-03, 3.7387207e-03,
          1.1531181e-03, 8.8139111e-04, 8.2804872e-05]
max_score = np.max(scores)  
scores = scores/max_score*100       
y_pos = np.arange(len(features))

axs.barh(y_pos, scores, height=0.45,left=5, align='center', color= '0.6', label=features)
axs.set_yticks(y_pos)
axs.set_yticklabels(labels=features,  fontsize= fs_y, color='0.1')

axs.invert_yaxis()  # labels read top-to-bottom

axs.set_xticks([0,25,50,75,100])
axs.set_xticklabels([0,0.25,0.50,0.75,1.0], fontsize= fs_ticks, color='0')
axs.set_title('Normal/Abnormal \n(cytoplasm)', fontsize=fs)

plt.show()
```

```python
## Grafico Binário (N + C)
 
 
#plt.rcdefaults()
fig, axs = plt.subplots(figsize=(1,3.8), sharey=True)
#plt.style.use('ggplot')
plt.style.use('seaborn')

fs = 11  # fontsize title
fs_ticks = 7  # fontsize  
fs_x = 10

plt.rcParams['font.family'] = 'Calibri'

#  data
features = ['hAreaN','eN','ratio_NC_per','ardN','ratio_NC_hArea','solidN','compacC',
            'ratio_NC','compacN','areaC','circuC','kN','circuN','periC','efdN50',
            'efdC69','periN','eC','equidiaN','raC','maxAxC','solidC','kC','efdC20',
            'elonC','efdN22','efdN6','efdC9','efdC1','sdnrlC']
scores = [0.11144678,0.09880449,0.09699471,0.09438305,0.09180647,0.09040274,
          0.08679786,0.08679099,0.07286223,0.07107463,0.07049827,0.0698288,
          0.0650465,0.059177,0.05817742,0.05428595,0.05369958,0.0487887,
          0.04153939,0.03885087,0.03860117,0.03851117,0.03746578,0.03681073,
          0.03627127,0.03352122,0.03306163,0.03228837,0.0315515,0.03054727]
max_score = np.max(scores)  
scores = scores/max_score*100   

features_N = ('hAreaN','eN','ardN','solidN','compacN','kN','circuN','efdN50',
            'periN','equidiaN','efdN22','efdN6')
scores_N = []
for f in features_N:
    scores_N.append(scores[features.index(f)])

f_N = ('convex hull','radial dist. entropy','aver. radial dist.','solidity','compacity',
              'kurtosis','circularity','EFD 50','perimeter','equi. diameter','EFD 22','EFD 6')
 
features_C = ('compacC','areaC','circuC','periC','efdC69','eC','raC','maxAxC','solidC','kC','efdC20',
            'elonC','efdC9','efdC1','sdnrlC') 
scores_C = []
for f in features_C:
    scores_C.append(scores[features.index(f)])

f_C = ('compacity','area','circularity','perimeter','EFD 69','radial dist. entropy',
       'aver. radial dist','major axis','solidity','kurosis','EFD 20', 'elongation',
       'EFD 9','EFD 1','std radial dist.') 
    
    
features_NC = ('ratio_NC_per','ratio_NC_hArea') 
scores_NC = []
for f in features_NC:
    scores_NC.append(scores[features.index(f)])
   
f_NC = ('perimeter ratio', 'convex hull ratio')

y_pos = np.arange(len(features_N))
axs.barh(y_pos, scores_N, height=0.4,left=5, align='center', color= '0.6')
axs.set_yticks(y_pos)
axs.set_yticklabels(labels=f_N,  fontsize= fs_y, color='0.1')

axs.invert_yaxis()  # labels read top-to-bottom

axs.set_xticks([0,25,50,75,100])
axs.set_xticklabels([0,0.25,0.50,0.75,1.0], fontsize= fs_ticks, color='0')
axs.set_title('Nucleus', fontsize=fs_x)

plt.show()
 
```

```python
fig, axs = plt.subplots(figsize=(1,3.8), sharey=True)
plt.rcdefaults()
plt.style.use('seaborn')
plt.rcParams['font.family'] = 'Calibri'

y_pos = np.arange(len(features_C))
axs.barh(y_pos, scores_C, height=0.5,left=5, align='center', color= '0.6')
axs.set_yticks(y_pos)
axs.set_yticklabels(labels=f_C, fontsize= fs_y, color='0.1')

axs.invert_yaxis()  # labels read top-to-bottom

axs.set_xticks([0,25,50,75,100])
axs.set_xticklabels([0,0.25,0.50,0.75,1.0], fontsize= fs_ticks, color='0')
axs.set_title('Cytoplasm', fontsize=fs)

plt.show()
```
